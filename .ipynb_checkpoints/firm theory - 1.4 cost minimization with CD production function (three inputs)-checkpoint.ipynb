# Solve for cost-minimizing input bundle analytically

## Setup the optimization program

The production manager's optimization problem with CD production function and three inputs, labor, capital and material, is
\begin{align}
    & \min_{L,K} \quad wL + rK + mM \\
    \text{s.t.}\quad & L^{\alpha} K^{\beta} M^{\gamma} = \overline{q}
\end{align}
where $m$ is the constant unit cost of material (M)

The Lagrangian is
\begin{align}
    \mathcal{L} = wL + rK + mM + \lambda ( \overline{q} - L^{\alpha} K^{\beta} M^{\gamma})
\end{align}

The FOCs are
\begin{align}
    w - \lambda \alpha L^{\alpha-1} K^{\beta} M^{\gamma} &= 0  \tag{1} \\
    r - \lambda \beta L^{\alpha} K^{\beta-1} M^{\gamma} &= 0 \tag{2} \\
    m - \lambda \gamma L^{\alpha} K^{\beta} M^{\gamma-1} &= 0 \tag{3} \\
    \overline{q} - L^{\alpha} K^{\beta} M^{\gamma} &= 0  \tag{4} 
\end{align}

Or, equivalently,
\begin{align}
    w &= \lambda \alpha L^{\alpha-1} K^{\beta} M^{\gamma}   \tag{1} \\
    r &= \lambda \beta L^{\alpha} K^{\beta-1} M^{\gamma}  \tag{2} \\
    m &= \lambda \gamma L^{\alpha} K^{\beta} M^{\gamma-1}  \tag{3} \\
    \overline{q} &= L^{\alpha} K^{\beta} M^{\gamma}   \tag{4} 
\end{align}

Dividing (1) by (2), we have
\begin{align}
    \frac{w}{r} &= \frac{\alpha}{\beta}\frac{K}{L}
\end{align}
or 
\begin{align}
    L &= \frac{\alpha}{\beta} \frac{r}{w} K \tag{5}
\end{align}

Dividing (3) by (2), we have
\begin{align}
    \frac{m}{r} &= \frac{\gamma}{\beta}\frac{K}{M}
\end{align}
or 
\begin{align}
    M &= \frac{\gamma}{\beta} \frac{r}{m} K \tag{6}
\end{align}

Note that both (5) and (6) are functions of K, we plug them into (4)
\begin{align}
    \overline{q} &= \left( \frac{\alpha}{\beta} \frac{r}{w} K \right)^{\alpha} K^{\beta} \left( \frac{\gamma}{\beta} \frac{r}{m} K  \right)^{\gamma} \\
    \overline{q} &= \left( \frac{\alpha}{\beta} \frac{r}{w} \right)^{\alpha} \left( \frac{\gamma}{\beta} \frac{r}{m}  \right)^{\gamma} K^{\alpha + \beta + \gamma}  \\
    K^{\alpha + \beta + \gamma} &= \left( \frac{\beta}{\alpha} \frac{w}{r} \right)^{\alpha} \left( \frac{\beta}{\gamma} \frac{m}{r}  \right)^{\gamma} \overline{q}  \\
    K^{\alpha + \beta + \gamma} &= \left( \frac{\beta}{r} \frac{w}{\alpha} \right)^{\alpha} \left( \frac{\beta}{r} \frac{m}{\gamma}  \right)^{\gamma} \overline{q}  \\
    K^{\alpha + \beta + \gamma} &= \left( \frac{\beta}{r} \right)^{\alpha + \gamma} \left( \frac{w}{\alpha} \right)^{\alpha} \left( \frac{m}{\gamma}  \right)^{\gamma} \overline{q}  \\
    K^{\alpha + \beta + \gamma} &= \left( \frac{\beta}{r} \frac{w}{\alpha} \right)^{\alpha} \left( \frac{\beta}{r} \frac{m}{\gamma}  \right)^{\gamma} \overline{q}  \\
    K^{\alpha + \beta + \gamma} &= \left( \frac{\beta}{r} \right)^{\alpha + \gamma} \left( \frac{w}{\alpha} \right)^{\alpha} \left( \frac{m}{\gamma}  \right)^{\gamma} \overline{q}  \\
    K &= \left[ \left( \frac{\beta}{r} \right)^{\alpha + \gamma} \left( \frac{w}{\alpha} \right)^{\alpha} \left( \frac{m}{\gamma}  \right)^{\gamma} \overline{q} \right]^{\frac{1}{\alpha + \beta + \gamma} } 
\end{align}

which is the optimal demand for $K$, and it can rewritten as 
\begin{align} 
    K^* &= \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }   \\
\end{align}

Plugging $K^*$ into (5), we can find the optimal demand for labor
\begin{align}
    L^* &= \frac{\alpha}{\beta} \frac{r}{w} K^* \\
        &= \frac{\alpha}{w} \frac{r}{\beta} K^* \\
        &= \left( \frac{w}{\alpha} \right)^{-1} \left( \frac{\beta}{r}  \right)^{-1} K^* \\
        &= \left( \frac{w}{\alpha} \right)^{-1} \left( \frac{\beta}{r}  \right)^{-1} \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } \\
        &= \left( \frac{w}{\alpha} \right)^{- \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{\beta}{r}  \right)^{- \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma}} \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } \\
        &= \left( \frac{\alpha}{w} \right)^{ \frac{\beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}  \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } \\
\end{align}

Plugging $K^*$ into (6), we can find the optimal demand for material
\begin{align}
    M^* &= \frac{\gamma}{\beta} \frac{r}{m} K^* \\
        &= \frac{r}{\beta} \frac{\gamma}{m} K^* \\
        &= \left( \frac{\beta}{r}  \right)^{-1} \left( \frac{m}{\gamma} \right)^{-1} K^* \\
        &= \left( \frac{\beta}{r}  \right)^{-1} \left( \frac{m}{\gamma} \right)^{-1} \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } \\
        &= \left( \frac{\beta}{r}  \right)^{- \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{- \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma}} \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } \\
        &= \left( \frac{\gamma}{m}  \right)^{ \frac{\alpha + \beta}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}   \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } \\
\end{align}

Plugging in the optimal demands into cost, we have the cost function
\begin{align}
    C &= wL^* + rK^* + mM^* \\
      &= w \left( \frac{\alpha}{w} \right)^{ \frac{\beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}  \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }  
      \\
      & \quad +  r \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }  
      \\
      & \quad +  m \left( \frac{\gamma}{m}  \right)^{ \frac{\alpha + \beta}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}   \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } 
      \\
    &= w \left( \frac{\alpha}{w} \right)^{ \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{\alpha}{w} \right)^{- \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}  \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }   
      \\
      & \quad +  r \left( \frac{\beta}{r} \right)^{ \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{\beta}{r} \right)^{ -\frac{\beta}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } 
      \\
      & \quad + m \left( \frac{\gamma}{m}  \right)^{ \frac{\alpha + \beta + \gamma}{\alpha + \beta + \gamma} } \left( \frac{\gamma}{m}  \right)^{ -\frac{\gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}   \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }  
      \\
    &= w \left( \frac{\alpha}{w} \right) \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}  \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }  
      \\
      & \quad +  r \left( \frac{\beta}{r} \right) \left( \frac{r}{\beta} \right)^{ \frac{\beta}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }  
      \\
      & \quad +  m \left( \frac{\gamma}{m}  \right) \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}   \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} } 
    \\
    &= (\alpha + \beta + \gamma) \left( \frac{w}{\alpha} \right)^{ \frac{\alpha}{\alpha + \beta + \gamma} } \left( \frac{r}{\beta}  \right)^{ \frac{ \beta }{\alpha + \beta + \gamma}}  \left( \frac{m}{\gamma}  \right)^{ \frac{\gamma}{\alpha + \beta + \gamma} } \overline{q}^{ \frac{1}{\alpha + \beta + \gamma} }
\end{align}

Note that the cost function is a function of input prices, output level and productivity parameters.

# Sympy solution

## Setup the utility function

```python
### Note that it is important here to further declare these variables to be nonzero and positive
#    Thanks to Abhi Agarwal for discovering this trick
alpha, beta, gamma = symbols('alpha, beta, gamma',nonzero=True, positive=True)


# Note: in sympy syntax: use double star ** for power
q = L**(alpha) * K**(beta) * M**(gamma)
# show the defined function
q
```

## Setup the lagrangian

```python
lam = symbols('lambda',nonzero=True, positive=True)
lagr = w*L + r*K + m*M + lam *(qbar - q)
lagr
```

Solve the first-order conditions. 

Note the syntax of the solve function, especially the use of the brackets:
solve( (eq1, eq2, eq3), var1, var2, var3)

The equations are by default treated as equal to 0 by the solver.

```python
foc_sol = solve( ( diff(lagr,L), diff(lagr,K), diff(lagr,M), diff(lagr,lam) ), [L,K,M,lam])
foc_sol
```

The solution is a list with one element, because our solution is unique.

```python
foc_sol[0]
```

To access the first element in our solution, which is a list by itself, we use the list index again.

```python
Lstar = foc_sol[0][0].simplify()
Lstar
```

```python
Kstar = foc_sol[0][1].simplify()
Kstar
```

```python
Mstar = foc_sol[0][2].simplify()
Mstar
```

Substitute optimal input demands into the cost, to calculate the minimum cost level necessary at the given exogenous variables. 

The resulting function is our **cost function**, $C(w,r,\overline{q})$.

```python
C = (w*L + r*K + m*M).subs([(L,Lstar), (K,Kstar), (M,Mstar)])
# simplify function does not work well here, but we can further simplify by hand
C.simplify()
```
